**Explique a relação do histograma com o boxplot, o que um mostra que o outro não? Como é possível inferir um do outro? Sugestão: plotar um boxplot e um histograma de um mesmo atributo pode ajudar a entender e responder a pergunta.**

O boxplot pode ser visto como uma aproximação do histograma, é possível ter uma ideia da distribuição pelo boxplot mas de maneira mais grosseira que pelo histograma.
Se uma distribuição for assimétrica, o boxplot mostrará as barras, q1 e q2, ou q2 e q3 mais próximas. 
Logo, tendo o histograma é possível ter uma ideia de como será o boxplot imaginando onde fica o q1, q2 e q3.


```python
# apenas plotando o histograma sem qualquer tratamento

columns = iris.columns.tolist()
plt.figure(figsize=(8, 10))
for col in range(1,5):
    ax = plt.subplot(2, 2, col)
    iris.iloc[:, col-1].plot.hist()
    ax.set_title(columns[col-1])

plt.tight_layout()
plt.show()
```

```python
# separando por classe o histograma de cada atributo

columns = iris.columns.tolist()

plt.figure(figsize=(10, 10))
for col in range(1,5):
    ax = plt.subplot(2, 2, col)
    ax.set_title(columns[col-1])
    for c in iris.target.unique():
        iris.loc[ iris.target == c, columns[col-1]].plot.hist(alpha=0.3)
    
plt.legend(target_names)
plt.tight_layout()
plt.show()
```

**Como você fez a escolha do parâmetro de intervalo (*bin*)? Por faixa (para altura, por exemplo, intervalo de 5cm), por quantidade (cada intervalo com tamanho 5, 0-5, 5-10, ...) ... ? Por que?**


**O parâmetro *bin* pode ser complicado para ajustar corretamente e pode distorcer bastante o histograma. Uma regra prática é tentar garantir que cada intervalo tenha um número mínimo de exemplos, uns 30 em cada *bin* deve ser suficiente. Mesmo assim nos *bins* das pontas é difícil cumprir essa regra.**

Utilizei o valor padrão para o número de bins, 10. Infelizmente, esse dataset não é grande o suficiente para ter essas garantias de um bom número de amostras por bin. Isso pode ser observado no último histograma do atributo petal width no qual algumas barras aparecem isoladas, isso é um indício de que os bins estão com poucos dados o histograma não tem nenhuma semelhança com nenhuma distribuição.

**Análise multivariada.**

**Muitas vezes a análise univariada, análise de cada atributo individualmente, não é muito informativa. Por isso também utilizamos a análise multivariada, nesta prática utilizaremos uma análise bivariada, analisando 2 atributos ao mesmo tempo.**


**3- Execute a covariância e a correlação entre os atributos selecionados.**
 

 



```python
cov = iris.iloc[:, [0,1,2,3]].cov()

sns.heatmap(cov)
plt.show()


```

```python
cor = iris.iloc[:, [0,1,2,3]].corr()

sns.heatmap(cor)
plt.show()
```

```python
# covarianca por classe
columns = iris.columns.tolist()

plt.figure(figsize=(10, 3))
for c_i, c in enumerate(iris.target.unique()):
    ax = plt.subplot(1, 3, c+1)
    plt.title(target_names[c_i])
    cov = iris.loc[ iris.target == c, columns[:-1]].cov()
    sns.heatmap(cov)
    
#plt.legend(target_names)
plt.tight_layout()
plt.show()
```

```python
# correlacao por classe
columns = iris.columns.tolist()

plt.figure(figsize=(10, 3))
for c_i, c in enumerate(iris.target.unique()):
    ax = plt.subplot(1, 3, c+1)
    plt.title(target_names[c_i])
    cor = iris.loc[ iris.target == c, columns[:-1]].corr()
    sns.heatmap(cor)
    
#plt.legend(target_names)
plt.tight_layout()
plt.show()
```

 **No seu caso, qual das duas é mais informativa? Explique.**
 
 **Qual a diferença entre as duas e quando é mais indicado utilizar uma ao invés da outra?**
 
 **De que forma elas ajudam, se é que ajudam, a responder suas perguntas?**
 
 
 
 De forma geral, se a variância de um dos atributos for muito maior que o de outro, isso afetará a covariancia e pode passar a impressão que existe uma relação entre as variáveis.
 Como a correlação corrige pelo desvio padrão de cada atributo, esse efeito é diminuído, mas ainda assim essa medida não captura tudo que creditamos a ela
 
 Apesar de ser parte a análise multivariada, a covariância e correlação não ajudam na resposta das perguntas aqui feitas. Mas se você estiver analisando muitas variáveis essa pode ser uma maneira rápida de identificar variáveis redundantes. Se duas variáveis tem alta correlação pode ser que a informação de uma esteja presente na outra, portanto, você remover uma das variáveis, mas o mais correto é uma análise mais cuidadosa nesses casos.
